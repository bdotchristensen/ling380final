```python
import requests
import pandas as pd
import time
```

```python
uncased = pd.read_csv('./uncased_finnish_nouns.csv', header = None)
uncased.columns = ['noun']
uncased
```

```python
finnish_cases = []

for i, x in enumerate(uncased.iterrows()):
    x = x[1][0]
    
    url = "https://cooljugator.com/fin/" + x
    #print(i, url)
    
    link = requests.get(url)
    t = link.text
    
    c = []
    
    if (t.count('<div class="meta-form">') == 29): # to filter out odd cases

        c.append(x)

        for j in range(29):

            # gen, dat, acc
            ii = t.find('<div class="meta-form">') + len('<div class="meta-form">')
            jj = t.find('</div><div class="meta-transliteration">')

            if j == 16: # accusative 1
                c.append(t[ii:jj])

            elif j == 17: # accusative 2
                c.append(t[ii:jj])

            elif j == 18: # genitive
                c.append(t[ii:jj])
                
            elif j == 19: # partitive
                c.append(t[ii:jj])
                
            elif j == 20: # inessive
                c.append(t[ii:jj])
                
            elif j == 21: # elative
                c.append(t[ii:jj])
                
            elif j == 22: # illative
                c.append(t[ii:jj])
                
            elif j == 23: # adessive
                c.append(t[ii:jj])
                
            elif j == 24: # ablative
                c.append(t[ii:jj])
                
            elif j == 25: # allative
                c.append(t[ii:jj])
                
            elif j == 26: # essive
                c.append(t[ii:jj])
                
            elif j == 27: # translative
                c.append(t[ii:jj])
                
            elif j == 28: # abessive
                c.append(t[ii:jj])
                

            t = t[(jj + 10):]
            
        finnish_cases.append(c)
            
            
    if i % 25 == 0:
        print(i, len(finnish_cases), c)  
    
    time.sleep(2.05)
    
```

```python
finnish_cases_df = pd.DataFrame(finnish_cases)
finnish_cases_df.columns = ['accusative_nom', 'accusative_gen', 'genitive', 'partitive', 'inessive', 'elative',
                           'elative', 'illative', 'adessive', 'ablative', 'allative', 'essive', 'translative',
                           'abessive']
finnish_cases_df
```

```python
finnish_cases_df.to_csv('./finnish_nouns_cased.csv')
```
