```python
import requests # make sure python 3.8 is running on this one
import pandas as pd
import time
```

```python
nouns = []
```

```python
link = requests.get("https://en.openrussian.org/list/nouns")
t = link.text

while t.find('"bare":"') != -1:
    i = t.find('"bare":"') + len('"bare":"')
    j = t.find('","accented"')
    
    nouns.append(t[i:j])
    #print(t[i:j])
    t = t[(j + 10):]

for i in range(50,2000,50):
    url = "https://en.openrussian.org/list/nouns?start=" + str(i)
    
    link = requests.get(url)
    t = link.text
    
    while t.find('"bare":"') != -1:
        i = t.find('"bare":"') + len('"bare":"')
        j = t.find('","accented"')

        nouns.append(t[i:j])
        #print(t[i:j])
        t = t[(j + 10):]
        
    time.sleep(2.1) # so it doesn't block me
    
```

```python
len(nouns)
```

```python
nouns_df = pd.DataFrame(nouns)
nouns_df.columns = ['uncased']
nouns_df
```

```python
nouns_df.to_csv('./nouns_russian_uncased.csv')
```

```python
russian_cases = []

for i, x in enumerate(nouns):
    url = "https://cooljugator.com/run/" + x
        
    link = requests.get(url)
    t = link.text
    
    c = []
    
    if (t.count('<div class="meta-form">') == 12): # to filter out odd cases

        c.append(x)

        for j in range(12):

            # gen, dat, acc
            ii = t.find('<div class="meta-form">') + len('<div class="meta-form">')
            jj = t.find('</div><div class="meta-transliteration">')

            if j == 7: # gen
                c.append(t[ii:jj])

            elif j == 8: # dat
                c.append(t[ii:jj])

            elif j == 9: # acc
                c.append(t[ii:jj])                          
                
            elif j == 10: # inst
                c.append(t[ii:jj])
                
            elif j == 11: # prep
                c.append(t[ii:jj])

            t = t[(jj + 10):]
            
        russian_cases.append(c)
            
            
    if i % 25 == 0:
        print(i, len(russian_cases), c)  
    
    time.sleep(2.05)
    
```

```python
russian_cases_df = pd.DataFrame(russian_cases)
russian_cases_df.columns = ['uncased', 'genitive', 'dative', 'accusative', 'instrumental', 'prepositional']
russian_cases_df
```
