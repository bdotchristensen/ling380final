```python
import pandas as pd
```

```python
allwords = []

with open('./german_10k.txt', 'rb') as f:
    allwords = (f.read()).split()
    
len(allwords)
```

```python
def is_noun(x):
    return x[0].isupper()
```

```python
nouns = []

for x in allwords:
    
    x = x.decode("utf-8")
    #print(x)
    
    if is_noun(str(x)):
        nouns.append(str(x))
        
len(nouns)
```

```python
nouns_df = pd.DataFrame(nouns)
nouns_df.columns = ['noun']
nouns_df
```

```python
all_nouns = pd.read_csv('./german_all_nouns.csv')
all_nouns
```

```python
nouns_wd = nouns_df.merge(all_nouns, how = 'inner', left_on = 'noun', right_on = 'lemma')
nouns_wd = nouns_wd.dropna()
nouns_wd
```
